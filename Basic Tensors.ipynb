# Tensorflow Constants

```
x = tf.constant([3,5,7])
x.shape, x.ndim, type(x), x
```

```
x = tf.constant([[3,5,7],
                 [4,6,8]])
x.shape, x.ndim, type(x), x
```

```
x.ndim
```

```
x = tf.constant([[3,5,7],
                 [4,6,8],
                 [1,3,5]])
x.shape, x.ndim, type(x), x
```

```
x = tf.constant([[[3,5,7],
                 [4,6,8],
                 [1,3,5],
                 [9,7,5]]])
x.shape, x.ndim, type(x), x
```

```
x1 = tf.constant([2,3,4])
print("x1:", x1.shape, x1.ndim, type(x1), x1)

x2 = tf.stack([x1,x1])
print("x2:", x2.shape, x2.ndim, type(x2), x2)

```

```
x3 = tf.stack([x2,x2,x2,x2])
print("x3:", x3.shape, x3.ndim, type(x3), x3)
```

```

x4 = tf.stack([x3,x3])
print("x4:", x4.shape, x4.ndim, type(x4), x4)
```

```
x = tf.constant([
                 [3,5,7],
                 [2,4,6]
])
print(x.shape)
#print(x[:,1])
y = x[1,:]
print("Y:",y)
```

```
z = tf.reshape(x, [3,2])
print("z", z)
```
